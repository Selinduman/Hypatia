```python
#Calculate and show the daily returns
daily_returns = stocks.pct_change(1)
daily_returns.dropna(inplace=True)
daily_returns
```

```python
#Calculate and show the annual returns
annual_returns = daily_returns.mean()*252
annual_returns
```

```python
#Calculate and show the annual risks(volatility)
annual_risks =daily_returns.std()*math.sqrt(252)
annual_risks
```

```python
#Description of daily returns
daily_returns_decs = daily_returns.describe()
daily_returns_decs
```

```python
plt.plot(daily_returns, alpha=.9);
plt.xlabel('Time')
plt.ylabel('Daily Returns')
```

```python
#Show annualize of assets
plt.bar(annual_returns.index,annual_returns)
plt.ylabel("Annualized Returns")
plt.xlabel("Stocks")
plt.xticks(rotation=90)
plt.title("Annualized Returns of Stocks")
plt.savefig("asfA-figure.png")
plt.show()
```

```python
#Create weights
n=1.0 / 3.0
weights = [n] * 3
weights = np.array(weights)
weights
```

```python
#Calculate and show the covariance matrix
cov_matrix_annual = daily_returns.cov()*252
cov_matrix_annual
```

```python
#Calculate and show the annual variance
port_variance=np.dot(weights.T, np.dot(cov_matrix_annual,weights))
port_variance
```

```python
#Calculate and show the portfolio annual volatility or risk
port_volatility = np.sqrt(port_variance)
port_volatility
```

```python
#Calculate and show the expected 
port_returns_expected =np.sum(weights*annual_returns)
port_returns_expected
```

```python
print(str(round(port_returns_expected*100,2))+"%")
```

```python
#Calculate portfolio daily returns
portfolio_weights = np.array([0.33333333, 0.33333333, 0.33333333])
port_ret = daily_returns.mul(portfolio_weights, axis=1).sum(axis=1)
port_ret
```

```python
from sklearn.decomposition import PCA
pca = PCA()
daily_returns_pca = pca.fit_transform(daily_returns_df)
```

```python
daily_returns_pca
```

```python
daily_returns_pca_df = pd.DataFrame(data = daily_returns_pca, columns=['PC1','PC2','PC3'])
```

```python
daily_returns_pca_df
```

```python
daily_returns_decs = daily_returns_pca_df.describe()
daily_returns_decs
```

```python
var_ratio=pca.explained_variance_ratio_
var_ratio=var_ratio*100
var_ratio_numpy=np.round(var_ratio,decimals = 2)
print(var_ratio)
```

```python
labels=["PC-" + str(x+1) for x in range(len(var_ratio))]
```

```python
plt.bar(labels, var_ratio)
plt.title("Explained Varience Based on PCA Components")
plt.xlabel("Components")
plt.ylabel("Explained Varience Ratio")
plt.savefig("PCA-figure.png")
plt.show()
```
